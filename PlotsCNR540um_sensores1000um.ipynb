```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#==========  Loading data  =============
```

```python
data=np.loadtxt('datosCNRCdTe1000um.txt')
data2=np.loadtxt('datosCNRSe1000um.txt')
data3=np.loadtxt('datosCNRGaAs1000um.txt')
data4=np.loadtxt('datosCNRMAPbI31000um.txt')
data5=np.loadtxt('datosCNRSi1000um.txt')
```

```python
#  Plot
```

```python
fig,ax=plt.subplots(figsize=(6,6))
sensors=['aSe','GaAs','Perovskite','CdTe','aSi']
bar_colors=['red','blue','green','olive','brown']
counts=[data2[4,2],data3[4,2],data4[4,2],data[4,2],data5[4,2]]
errcounts=[data2[4,3],data3[4,3],data4[4,3],data[4,3],data5[4,3]]
hbar=ax.bar(sensors,counts,yerr=errcounts,color=bar_colors)
ax.bar_label(hbar,fmt='%.2f')
ax.set_ylabel('CNR')
ax.set(ylim=(0,20))
ax.text(0.45,0.95,'Sphere 540 $\mu m$',transform=ax.transAxes,fontsize=10)
ax.text(0.45,0.91,'Thickness of the sensors: 1000 $\mu m$',transform=ax.transAxes,fontsize=10)
plt.show()
```
